```python
import selenium
from selenium import webdriver
import time
import pandas as pd
from random import randint
```

```python
adat = pd.read_csv('top_25000.csv')
elso_adag = adat['Name'][6:200]
my_links = 'https://steemit.com/'+'@'+elso_adag
```

```python

driver = webdriver.Chrome() 
driver.get('https://steemit.com/')
time.sleep(30)
for i in my_links:
    print(i)
    driver.get(i)
    driver.find_element_by_xpath('//*[@id="content"]/div/div/div[2]/div/div[1]/div/div[1]/div/span/label[1]').click()
    time.sleep(randint(5, 9))
```
